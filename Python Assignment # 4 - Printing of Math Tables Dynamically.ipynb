```python
print("******You can Print Tables of your choice******")
table = int(input("Enter a Range to which you can generate table:"))
print("\n")
for e in range(1,table+1):
    print(f'Table of {e} \n')
    for i in range(1,11):
        print(f'{e} x {i} = {i*e}')
    print(f'***********\n')
        
```
