```python
#Object 2 Dynamically Print All list() Attributes

l1 = []
s1 = ""
t1 = ()
d1 = {}


print("*******************ATTRIBUTES OF STRING*******************\n")

for j in dir(s1)[:34]:
    print(getattr(s1,j))
print("\n*******************ATTRIBUTES OF LIST*******************\n")

for i in dir(l1)[:35]:
    print(getattr(l1,i))
print("\n*******************ATTRIBUTES OF TUPLE*******************\n")

for k in dir(t1)[:32]:
    print(getattr(t1,k))
print("\n*******************ATTRIBUTES OF DICTIONARY*******************\n")

for l in dir(d1)[:30]:
    print(getattr(d1,l))
```

```python
t1 = ("Pakistan","China","India","Pakistan")
print(t1.count("Pakistan")) #Returns the number of times a specified value occurs in a tuple
print(t1.index("India")) #Searches the tuple for a specified value and returns the position of where it was found
```

```python
d1 = {
    "Name":"Basit",
    "Age": "20 Years",
    "Class":"PIAIC"    
}
t1 = ("Pakistan","China","India","Pakistan")
t_index = 0

print(d1.copy()) #Returns a copy of the dictionary
print(d1.fromkeys(t1,t_index)) #Returns a dictionary with the specified keys and values
print(d1.get("Age"))  #Returns the value of the specified key
print(d1.items()) #Returns a list containing the a tuple for each key value pair
print(d1.keys()) #Returns a list containing the dictionary's keys
print(d1.pop("Age")) #Removes the element with the specified key
print(d1.popitem()) #Removes the last inserted key-value pair
print(d1.setdefault("Class")) #Returns the value of the specified key. If the key does not exist: insert the key, with the specified value
d1.update({"color": "White"}) #Updates the dictionary with the specified key-value pairs
print(d1.values()) #Returns a list of all the values in the dictionary
d1.clear() #Removes all the elements from the dictionary
print(d1)
```
