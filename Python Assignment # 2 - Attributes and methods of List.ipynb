```python
#Object 1 to use all list() Methods

l1 = list(map(chr,list(range(65,75))))
l2 = ["Extended Element 1","Extended Element 2","Extended Element 3"]
print(l1)

l1.append("Appended Element") #Adds an element at the end of the list
print(l1.copy())     #Returns a copy of the list
print(l1.count("D"))    #Returns the number of elements with the specified value
l1.extend(l2)   #Add the elements of a list (or any iterable), to the end of the current list
print(l1) 
print(l1.index("F"))   #Returns the index of the first element with the specified value
l1.insert(7,"Inserted Element")   #Adds an element at the specified position
print(l1.pop())     #Removes the element at the specified position
l1.remove("Inserted Element")   #Removes the first item with the specified value
l1.reverse()  #Reverses the order of the list
print(l1) 
l1.sort()    #Sorts the list
print(l1) 
l1.clear()    #Removes all the elements from the list
print(l1)
```

```python
#Object 2 Dynamically Print All list() Attributes

l1 = list(range(1,100))
List_attr = dir(l1)[:35]
[getattr(l1,e) for e in List_attr]
```

```python
#Object 3 Use list comprehension with Loop and If Else

l1 = list(range(1,1001))
[x if x%5 == 0 else "Not Found at this index" for x in l1]
```
